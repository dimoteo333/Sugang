```python
def download(url, params=None, headers={}, method='GET'):
    req = request.Request(url + '?' + (parse.urlencode(params) 
                          if method == 'GET' else ''),
                          data = parse.urlencode(params).encode()
                          if method == 'POST' else None,
                          headers=headers, method=method)
    
    resp = None
    
    try:
        resp = request.urlopen(req)
    except HTTPError as e:
        print(e.getcode(), e.reason)
        print(e.getheaders())
        
    return resp
```

```python
from requests import request, HTTPError
from requests.compat import urljoin, urlparse, quote, unquote
from time import sleep

def download(url, params={}, headers={}, method='GET', maximum=2):
    resp = request(url=url, params=params if method=='GET' else {},
            data=params if method=='POST' else {},
            headers=headers, method=method)
    
    
    try:
        resp.raise_for_status()
    except HTTPError as e:
        if 500 <= e.response.status_code < 600 and maximum > 0:
            print('retry')
            sleep(1) # 1초?
            resp = download(url, params, headers, method, maximum-1) # inf loop
        else:
            print(e)
        
    return resp
```

```python
# Crawler
# HTTP를 통해 Text (Markup Language), 어플리케이션의 JSON, XML들을 

# 웹을 따라 자동으로 탐험하는 봇
# 링크 정보 수집이 제일 중요 ... 링크를 통해 페이지들 간의 관계를 알아낼 수 있음.
# 목적 : Web Indexing (전문으로부터 어떤 내용이 있었는지 키워드 형태로 추려내는 과정.)
# 숫자.. 정량화... 모델링에 활용 가능.

# Crawling Architecture
# 하이퍼링크를 탐색하고, 찾고, 다시 fetch..하는 일련의 과정
# Seed URL이 필요함. (예) 수동으로 입력한 검색한 주소)

# 구체적인 과정
# REQUEST -> WEB SERVER -> HTML -> LINK EXTRACTION -> URL LIST -> SEED URL -> REQUEST ...

# 하기 전에..
# 공공 데이터 API 사용법
# 개인 키 승인이 늦어지고 있음..
```

```python
# 구조를 쭉 따라 온거임. 
for _ in dom.find('table').find_all(recursive=False): # => 모든 행
    if _.find('td'):
        print(_.find_all('td')[2].text.strip())
    else:
        print(_.find('th').text.strip())
```
